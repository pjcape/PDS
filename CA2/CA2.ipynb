# Nature Of The Dataset:

The CSV file "graduate-employment-survey-ntu-nus-sit-smu-suss-sutd" contains data on the employment outcomes of graduates from the six autonomous universities in Singapore (NUS, NTU, SMU, SUTD, SIT, and SUSS). The data includes the following columns:

- Year: The year in which the graduate graduated.
- University: The university from which the graduate graduated.
- School: The school within the university from which the graduate graduated.
- Degree: The degree that the graduate graduated with.
- Employment_rate_overall: The overall employment rate of graduates from the university.
- Employment_rate_ft_perm: The full-time permanent employment rate of graduates from the university.
- Basic_monthly_mean: The mean basic monthly salary of graduates from the university.
- Basic_monthly_median: The median basic monthly salary of graduates from the university.
- Gross_monthly_mean: The mean gross monthly salary of graduates from the university.
- Gross_monthly_median: The median gross monthly salary of graduates from the university.
- Gross_mthly_25_percentile: The 25th percentile of the gross monthly salaries of graduates from the university.
- Gross_mthly_75_percentile: The 75th percentile of the gross monthly salaries of graduates from the university.

The dataset contains a total of 1121 rows of data. The data is relatively clean and well-formatted. However, there are a few peculiarities that should be noted.

- The data does not include information on the gender of the graduates. This information would be useful for understanding the differences in employment outcomes between male and female graduates.
- The data does not include information on the race of the graduates. This information would also be useful for understanding the differences in employment outcomes between different racial groups.
- The data does not include information on the undergraduate GPA of the graduates. This information would be useful for understanding the relationship between undergraduate GPA and employment outcomes.
- It is also to be noted that the number of graduates surveyed for each university is different, NTU & NUS graduates accounted for most of the surveys conducted.

I analyzed the dataset using the following steps:

1. I imported the data into a Python Pandas DataFrame.
2. I performed a statistical analysis of the data, including calculating the mean, range of the employment rates, basic monthly salaries.
3. I created visualizations of the data, such as line charts, bar charts, box plots, and scatterplots.

# What is the employment rates of graduates from various universities in Singapore?

```python
# Clean the data and convert numeric columns to appropriate data types
numeric_columns = ['employment_rate_overall', 'employment_rate_ft_perm', 'basic_monthly_mean', 'basic_monthly_median', 'gross_monthly_mean', 'gross_monthly_median', 'gross_mthly_25_percentile', 'gross_mthly_75_percentile']
graduate_employment_survey_ntu_nus_sit_smu_suss_sutd[numeric_columns] = graduate_employment_survey_ntu_nus_sit_smu_suss_sutd[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Drop rows with NaN values in the 'employment_rate_ft_perm' column
cleaned_data_ft = graduate_employment_survey_ntu_nus_sit_smu_suss_sutd.dropna(subset=['employment_rate_ft_perm'])

# Group the data by university and calculate the average employment rate for each university
university_employment_rates_2018 = cleaned_data_ft.loc[cleaned_data_ft['year'] == 2018].groupby('university')['employment_rate_ft_perm'].mean()
university_employment_rates_2021 = cleaned_data_ft.loc[cleaned_data_ft['year'] == 2021].groupby('university')['employment_rate_ft_perm'].mean()

# Plot the bar graph
plt.figure(figsize=(14, 14))

# full-time data for 2018 
plt.subplot(2, 1, 1)
bars = plt.bar(university_employment_rates_2018.index, university_employment_rates_2018.values)
plt.xlabel('University')
plt.ylabel('Average Employment Rate (%)')
plt.title('Average Full-Time(Permanent) Employment Rates of Graduates by University in Singapore(2018)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid()
# Add the values of each bar to the corresponding bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}%',
             ha='center', va='bottom', fontsize=10, color='black')

# full-time data for 2021
plt.subplot(2, 1, 2)
bars = plt.bar(university_employment_rates_2021.index, university_employment_rates_2021.values)
plt.xlabel('University')
plt.ylabel('Average Employment Rate (%)')
plt.title('Average Full-Time(Permanent) Employment Rates of Graduates by University in Singapore(2021)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid()
# Add the values of each bar to the corresponding bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}%',
             ha='center', va='bottom', fontsize=10, color='black')
plt.show()
```

From the bar graph, it shows the Average Full-Time(Permanent) Employment Rates of Graduates by University in Singapore 2018 & 2021. The highest average employment rate for a University in 2018 would be Singapore University of Technology and Design (SUTD) 89.53% whereas in 2021 would be Singapore Management University(SMU) 90.43%. The lowest average employment rate for a University in 2013 would be Nanyang Technological University(NTU) 79.01 % whereas in 2021 would be Singapore University of Social Sciences(SUSS) 78.55%. It can also be noted that all universites have increased employment rates from 2013 to 2021, with the exception of SUSS with an employment rate of 88.50% in 2013 and 78.55% in 2021. 

# What is the trend of the mean basic monthly salary over the years for local universities in Sinagpore ?

```python
# Clean the data and convert numeric columns to appropriate data types
numeric_columns = ['employment_rate_overall', 'employment_rate_ft_perm', 'basic_monthly_mean', 'basic_monthly_median', 'gross_monthly_mean', 'gross_monthly_median', 'gross_mthly_25_percentile', 'gross_mthly_75_percentile']
graduate_employment_survey_ntu_nus_sit_smu_suss_sutd[numeric_columns] = graduate_employment_survey_ntu_nus_sit_smu_suss_sutd[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Drop rows with NaN values in the 'basic_monthly_mean' column
cleaned_data = graduate_employment_survey_ntu_nus_sit_smu_suss_sutd.dropna(subset=['basic_monthly_mean'])

# Group the data by university and year, and calculate the mean salary for each university in each year
grouped_data = cleaned_data.groupby(["university", "year"])["basic_monthly_mean"].mean().reset_index()

# Create a line chart to show the trend of mean salaries over the years for each university
plt.figure(figsize=(12, 6))

universities = grouped_data["university"].unique()

for university in universities:
    data_by_university = grouped_data[grouped_data["university"] == university]
    plt.plot(data_by_university["year"], data_by_university["basic_monthly_mean"], label=university)

plt.xlabel('Year')
plt.ylabel('Mean Basic Monthly Salary(SGD)')
plt.title('Trend of Mean Basic Monthly Salary Over the Years for Each University')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.tight_layout()
plt.show()
```

From the line chart, it shows the Trend of Mean Basic Monthly Salary Over the Years for Each University. It is evident that there is a notable relationship between the passage of time and the increase in mean basic monthly salary for all universities. There is also a steep increase in basic monthly salary between 2018-2019 for NUS(around SGD3700 to around SGD4000), 2020-2021 for SMU(around SGD4300 to around SGD4600) & SUSS(around SGD3200 to around SGD3450). In 2021, the lowest basic monthly salary would be around SGD3450 for SUSS and the highest would be be around SGD4600 for SMU.

# How does the mean basic monthly salary correlate with the employment rate?

```python
# Clean the data and convert numeric columns to appropriate data types
numeric_columns = ['employment_rate_overall', 'employment_rate_ft_perm', 'basic_monthly_mean', 'basic_monthly_median', 'gross_monthly_mean', 'gross_monthly_median', 'gross_mthly_25_percentile', 'gross_mthly_75_percentile']
graduate_employment_survey_ntu_nus_sit_smu_suss_sutd[numeric_columns] = graduate_employment_survey_ntu_nus_sit_smu_suss_sutd[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Drop rows with NaN values in the 'basic_monthly_mean' & 'employment_rate_ft_perm' column
cleaned_data = graduate_employment_survey_ntu_nus_sit_smu_suss_sutd.dropna(subset=['basic_monthly_mean', 'employment_rate_ft_perm'])

# Basic monthly salary vs. employment rate
plt.figure(figsize=(12, 6))
plt.scatter(cleaned_data['employment_rate_ft_perm'], cleaned_data['basic_monthly_mean'], marker='o', c='b', label='Data Points')

# Calculate the best-fit line (linear regression)
coefficients = np.polyfit(cleaned_data['employment_rate_ft_perm'], cleaned_data['basic_monthly_mean'], 1)
poly_function = np.poly1d(coefficients)
x_values = cleaned_data['employment_rate_ft_perm']
y_fit = poly_function(x_values)

# Plot the best-fit line
plt.plot(x_values, y_fit, color='r', label='Best Fit Line')

plt.xlabel('Full-Time Employment Rate (%)')
plt.ylabel('Basic Monthly Salary(SGD)')
plt.title('Full-Time Employment Rate vs. Basic Monthly Salary (2013-2021)')
plt.legend()
plt.grid(True)
plt.show()
```

From the scatterplot, it shows the relationship between basic monthly salary and the full-time employment rate . It is the highest at 100% employment rate with just around SGD3800. This is because when everyone who graduates is employed full-time, the demand for labor is high, which drives up wages. It is the lowest at around 30% employment rate with just around SGD2900. This is because when there are fewer jobs available, the competition for those jobs is fierce, which drives down wages. It is evident that there is a positive correlation between the full-time employment rate and the basic monthly salary. As the full-time employment rate increases, the basic monthly salary also tends to increase.

When the demand for labor is high, the basic monthly salary is high. When the demand for labor is low, the basic monthly salary is low.
