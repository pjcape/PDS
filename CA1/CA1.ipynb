# Nature Of The Dataset:

The CSV file "resale-flat-prices-based-on-registration-date-from-jan-2017-onwards" contains data on resale flat transactions in Singapore from January 2017 to the present. The data includes the following columns:

- Month: The month in which the resale transaction took place.
- Town: The town in which the resale flat is located.
- Flat type: The type of resale flat, such as 3-room, 4-room, 5-room etc.
- Block: The block number of the resale flat.
- Street name: The street name of the resale flat.
- Storey range: The storey range of the resale flat, such as 1-10, 11-20, 21-30 etc.
- Floor area (sqm): The floor area of the resale flat in square meters.
- Flat model: The model of the resale flat.
- Lease commence date: The date on which the lease on the resale flat commenced.
- Remaining lease: The number of years, months, and days of lease remaining on the resale flat.
- Resale price: The price at which the resale flat was sold.

The dataset contains a total of 154511 rows of data. The data is relatively clean and well-formatted. However, there are a few peculiarities that should be noted.

- The data does not include information on the age of the resale flat. This information would be useful for understanding the relationship between the age of a flat and its resale price.
- The data does not include information on the number of bedrooms and bathrooms in the resale flat. This information would also be useful for understanding the relationship between the number of bedrooms and bathrooms in a flat and its resale price.
- The data does not include information on the amenities and facilities that are located near the resale flat. This information would be useful for understanding the relationship between the amenities and facilities near a flat and its resale price.

# What is the distribution of the resale prices across different flat types?

```python
# Extract flat type and average resale price
flat_types = np.unique(resale_flat_prices['flat_type'])
avg_resale_price = []
for flat_type in flat_types:
    avg_resale_price.append(np.mean(resale_flat_prices[resale_flat_prices['flat_type']==flat_type]['resale_price']))

# Plotting the graph
plt.figure(figsize=(10, 4))
plt.bar(flat_types, avg_resale_price)
plt.xlabel('Flat Type')
plt.ylabel('Average Resale Price ($)')
plt.title('Distribution of Resale Prices Across Different Flat Types')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

From the bar graph, it shows the Distribution of Resale Prices Across Different Flat Types. The highest average resale price for a flat would be multi-generation flats. The lowest average resale price for a flat would be 1 room flats. It is evident that there is a clear relationship between the number of rooms in different flat types and the average resale price. As the number of rooms increases, there is a corresponding upward trend in the average resale price.

It can also be seen that the median average resale price would be that of a 4 room flat hence a considerable option for purchasing a flat.

## Which flat type is sold the most?

```python
room1_count = np.sum(resale_flat_prices['flat_type']=='1 ROOM')
room2_count = np.sum(resale_flat_prices['flat_type']=='2 ROOM')
room3_count = np.sum(resale_flat_prices['flat_type']=='3 ROOM')
room4_count = np.sum(resale_flat_prices['flat_type']=='4 ROOM')
room5_count = np.sum(resale_flat_prices['flat_type']=='5 ROOM')
executive_count = np.sum(resale_flat_prices['flat_type']=='EXECUTIVE')
mg_count = np.sum(resale_flat_prices['flat_type']=='MULTI-GENERATION')

room12mg_count = room1_count + mg_count + room2_count

sizes = [room3_count, room4_count, room5_count, executive_count, room12mg_count]

flat_type = np.unique(resale_flat_prices['flat_type'])

new_labels = []
for flat in flat_type:
    if flat == '1 ROOM':
        np.delete(flat, np.where(flat == '1 ROOM'))
    elif flat == '2 ROOM':
        np.delete(flat, np.where(flat == '2 ROOM'))
    elif flat == 'MULTI-GENERATION':
        continue
    else:
        new_labels.append(flat)
new_labels.append('1 ROOM + 2 ROOM + MULTI-GENERATION')

fig1,ax1 = plt.subplots(figsize=(8,8))
ax1.pie(sizes, autopct='%1.1f%%', textprops={'fontsize': 13})
plt.legend(new_labels,title="Flat Types",loc="center left",bbox_to_anchor=(1, 0.5))
plt.title("Distribution of Flat Types",fontsize=20)
plt.show()
```

From the pie chart, it shows the Distribution of Resale Flat Types. The majority of resale flats sold are 4 room flats with 41.9%. The least number of resale flats sold is the combination of 1 room, 2 room and multi generation flats all of which cumulatively account for 1.7%.
