### Function Declaration
The function below allows to generate a boolean array based on a list of values. It is used to extract from the dataset that matches the list of countries given.

```python
# Allows to generate a boolean array based on a list of values
def generate_array_bool(df: pd.DataFrame, col: str, values: list):
    return np.logical_or.reduce(
        np.array(
            [
            df[col] == val
            for val in values
            ]
        )
    )
```

# Annual Ranking of the Top 3 Countries by Happiness Index from 2018 to 2022
The Happiness Index is a collection of measurements that takes into account several contributing factors, including Social Support, Healthy Life Expectancy, Personal Freedom, Generosity, Perceptions of Corruption, and Gross Domestic Product per capita. The scores of these individual factors are combined to form an overall score for the Happiness Index.

```python
df_world_happiness_2022['Happiness score'] = df_world_happiness_2022['Happiness score'].str.replace(",", "").astype(float) / 1000


# list_of_countries_happiest = ["Finland", "Norway", "Denmark", "Iceland", "Switzerland", "Netherlands"]

# happiness_cond_2018 = df_world_happiness_2018[generate_array_bool(df_world_happiness_2018, 'Country or region', list_of_countries_happiest)]
# happiness_cond_2019 = df_world_happiness_2019[generate_array_bool(df_world_happiness_2019, 'Country or region', list_of_countries_happiest)]
# happiness_cond_2020 = df_world_happiness_2020[generate_array_bool(df_world_happiness_2020, 'Country name', list_of_countries_happiest)]
# happiness_cond_2021 = df_world_happiness_2021[generate_array_bool(df_world_happiness_2021, 'Country name', list_of_countries_happiest)]
# happiness_cond_2022 = df_world_happiness_2022[generate_array_bool(df_world_happiness_2022, 'Country', list_of_countries_happiest)]

happiness_cond_2018 = df_world_happiness_2018.head(3)
happiness_cond_2019 = df_world_happiness_2019.head(3)
happiness_cond_2020 = df_world_happiness_2020.head(3)
happiness_cond_2021 = df_world_happiness_2021.head(3)
happiness_cond_2022 = df_world_happiness_2022.head(3)

happiness_cond_2022 = happiness_cond_2022.filter(['Country', 'Happiness score'])
happiness_cond_2022['year'] = 2022
happiness_cond_2022.set_index('year', inplace=True)
happiness_cond_2022.rename(columns={'Country': 'country', 'Happiness score': 'score'}, inplace=True)

happiness_cond_2021 = happiness_cond_2021.filter(['Country name', 'Ladder score'])
happiness_cond_2021['year'] = 2021
happiness_cond_2021.set_index('year', inplace=True)
happiness_cond_2021.rename(columns={'Country name': 'country', 'Ladder score': 'score'}, inplace=True)

happiness_cond_2020 = happiness_cond_2020.filter(['Country name', 'Ladder score'])
happiness_cond_2020['year'] = 2020
happiness_cond_2020.set_index('year', inplace=True)
happiness_cond_2020.rename(columns={'Country name': 'country', 'Ladder score': 'score'}, inplace=True)

happiness_cond_2019 = happiness_cond_2019.filter(['Country or region', 'Score'])
happiness_cond_2019['year'] = 2019
happiness_cond_2019.set_index('year', inplace=True)
happiness_cond_2019.rename(columns={'Country or region': 'country', 'Score': 'score'}, inplace=True)

happiness_cond_2018 = happiness_cond_2018.filter(['Country or region', 'Score'])
happiness_cond_2018['year'] = 2018
happiness_cond_2018.set_index('year', inplace=True)
happiness_cond_2018.rename(columns={'Country or region': 'country', 'Score': 'score'}, inplace=True)

# Concat all of them together
happiness_cond = pd.concat([happiness_cond_2018.sort_values("score"), happiness_cond_2019.sort_values("score"), happiness_cond_2020.sort_values("score"), happiness_cond_2021.sort_values("score"), happiness_cond_2022.sort_values("score")])
happiness_cond['year'] = happiness_cond.index

# happiness_cond = happiness_cond.reset_index(drop=True)

# Plot line graph by year, all on the same graph
plt.figure(figsize=(15, 5))
plt.suptitle("Annual Ranking of the Top 3 Countries by Happiness Index from 2018 to 2022", fontsize=16)
plt.title("The higher the score, the happier the country")
plt.xlabel("Year")
plt.ylabel("Happiness score")
sns.barplot(data=happiness_cond, x='year', y='score', hue='country')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

## Heatmap of percentage of Secondary school students in Singapore from NT, NA and O Levels that passed Math every year

```python
df_new = df_math.groupby(['race', 'year']).agg({'percentage_pass_math': 'mean'}).reset_index()
df_new = df_new[df_new['year'] >= 2008]
df_new_without_overall = df_new[df_new['race'] != 'Overall']
df_new_without_overall.index = df_new_without_overall['year']
heatmap_data= df_new_without_overall.pivot("race", "year", "percentage_pass_math")

# Label the x and y axis
# Increase size of the plot
plt.gcf().set_size_inches(25, 10)
plt.ylabel('Race')
plt.xlabel('Year')
plt.title('Annual Percentage of Secondary School Students in Singapore Achieving a Passing Grade in Mathematics by Racial Group')
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap="YlGnBu")
```

### Insights/Remarks
-   The worst percentage overall is Malay students in 2008.
-   The best performing overall is Chinese students in 2021.
-   There is an increase from 2011 to 2012 for Chinese, Indians and Malays. However a decrease for Others.
